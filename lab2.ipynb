```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 12))

t = np.linspace(0, 2, 1000)
n = np.arange(0, 2, 0.1)

f = 2

sine_c = np.sin(2 * np.pi * f * t)
sine_d = np.sin(2 * np.pi * f * n)

plt.subplot(4, 2, 1)
plt.plot(t, sine_c, 'b')
plt.title('Continuous Sine Wave')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(4, 2, 2)
plt.stem(n, sine_d)
plt.title('Discrete Sine Wave')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 12))

t = np.linspace(-2, 2, 1000)
n = np.arange(-10, 11)

step_c = np.heaviside(t, 0.5)
step_d = np.heaviside(n, 0.5)

plt.subplot(4, 2, 1)
plt.plot(t, step_c, 'g-', linewidth=2)
plt.title('Continuous Unit Step Signal')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.ylim(-0.2, 1.2)
plt.grid(True)

plt.subplot(4, 2, 2)
plt.stem(n, step_d)
plt.title('Discrete Unit Step Signal')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.ylim(-0.2, 1.2)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 12))

t = np.linspace(-5, 5, 1000)
n = np.arange(-10, 11)

ramp_c = np.where(t >= 0, t, 0)
ramp_d = np.where(n >= 0, n, 0)

plt.subplot(4, 2, 1)
plt.plot(t, ramp_c, 'b')
plt.title('Continuous Ramp Signal')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(4, 2, 2)
plt.stem(n, ramp_d)
plt.title('Discrete Ramp Signal')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 12))
signum_c = np.sign(t)
signum_d = np.sign(n)

plt.subplot(4, 2, 3)
plt.plot(t, signum_c, 'g')
plt.title('Continuous Signum Signal')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(4, 2, 4)
plt.stem(n, signum_d)
plt.title('Discrete Signum Signal')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 12))
impulse_c = np.zeros_like(t)
impulse_c[np.abs(t) < 0.01] = 1

impulse_d = np.where(n == 0, 1, 0)

plt.subplot(4, 2, 5)
plt.plot(t, impulse_c, 'r')
plt.title('Continuous Unit Impulse Signal')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(4, 2, 6)
plt.stem(n, impulse_d)
plt.title('Discrete Unit Impulse Signal')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 12))
exp_c = np.exp(t)
exp_d = np.exp(n / 5)

plt.subplot(4, 2, 7)
plt.plot(t, exp_c, 'm')
plt.title('Continuous Exponential Signal')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(4, 2, 8)
plt.stem(n, exp_d)
plt.title('Discrete Exponential Signal')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
