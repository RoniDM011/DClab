```python
#  Conditional statements

# Simple if
temperature = 25
print(f"Temperature: {temperature}°C")

if temperature > 30:
    print("It's hot outside! 🥵")
elif temperature > 20:
    print("It's warm outside! 😊")
elif temperature > 10:
    print("It's cool outside! 🧥")
else:
    print("It's cold outside! 🥶")

# Nested conditions
age = 18
has_license = True

if age >= 18:
    if has_license:
        print("You can drive a car! 🚗")
    else:
        print("You're old enough but need a license first!")
else:
    print("You're too young to drive!")
```

```python
# Loops

# FOR loop - iterating over a range
print("For loop - Counting 1 to 5:")
for i in range(1, 6):
    print(f"Count: {i}")

# FOR loop - iterating over a list
print("\nFor loop - Iterating through fruits:")
fruits = ["apple", "banana", "orange", "grape"]
for fruit in fruits:
    print(f"I like {fruit}")

# FOR loop - iterating over a string
print("\nFor loop - Iterating through characters:")
word = "Python"
for letter in word:
    print(f"Letter: {letter}")

# WHILE loop
print("\nWhile loop - Countdown:")
count = 5
while count > 0:
    print(f"Countdown: {count}")
    count -= 1  # Decrease count by 1
print("Blast off! 🚀")

# Break and continue
print("\nBreak and Continue Example:")
for num in range(1, 11):
    if num == 5:
        continue  # Skip 5
    if num == 8:
        break     # Stop at 8
    print(f"Number: {num}")
```

```python
# Lists (Arrays)

# Creating lists
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]
colors = ["red", "green", "blue"]

print(f"Numbers: {numbers}")
print(f"Mixed list: {mixed}")
print(f"Colors: {colors}")

# Accessing elements
print(f"\nFirst color: {colors[0]}")
print(f"Last color: {colors[-1]}")

# List operations
colors.append("yellow")
print(f"After append: {colors}")

colors.insert(1, "orange")
print(f"After insert: {colors}")

colors.remove("green")
print(f"After remove: {colors}")

# Slicing lists
print(f"\nFirst 2 colors: {colors[:2]}")
print(f"Last 2 colors: {colors[-2:]}")

# List comprehension
squares = [x**2 for x in range(1, 6)]
print(f"\nSquares: {squares}")
```

```python
# DICTIONARIES

# Creating a dictionary
person = {
    "name": "John",
    "age": 30,
    "city": "New York",
    "is_student": False
}

print(f"Person dictionary: {person}")

# Accessing values
print(f"\nName: {person['name']}")
print(f"Age: {person.get('age')}")

# Modifying dictionary
person["age"] = 31
person["job"] = "Engineer"
print(f"\nUpdated person: {person}")

# Iterating through dictionary
print("\nIterating through dictionary:")
for key, value in person.items():
    print(f"{key}: {value}")
```

```python
# Functions

# Simple function
def greet():
    print("Hello! Welcome to Python!")

# Function with parameters
def greet_person(name):
    print(f"Hello, {name}!")

# Function with return value
def add_numbers(a, b):
    return a + b

# Function with default parameter
def multiply(a, b=2):
    return a * b

# Calling functions
greet()
greet_person("Alice")

sum_result = add_numbers(5, 3)
print(f"5 + 3 = {sum_result}")

product = multiply(5)
print(f"5 * 2 = {product}")

product2 = multiply(5, 4)
print(f"5 * 4 = {product2}")
```

```python
#Exception handling

# Handling division by zero
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Cannot divide by zero!"
    except Exception as e:
        return f"An error occurred: {e}"

print(f"10 / 2 = {safe_divide(10, 2)}")
print(f"10 / 0 = {safe_divide(10, 0)}")

# Handling invalid input
def get_number():
    try:
        num = int(input("Enter a number: "))
        print(f"You entered: {num}")
    except ValueError:
        print("That's not a valid number!")
get_number()
```

```python
#Practice examples

#Even or Odd checker
def check_even_odd():
    num = int(input("\nEnter a number to check if it's even or odd: "))
    if num % 2 == 0:
        print(f"{num} is EVEN")
    else:
        print(f"{num} is ODD")
check_even_odd()

#Sum of numbers in a list
def sum_list(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

my_numbers = [10, 20, 30, 40, 50]
print(f"\nList: {my_numbers}")
print(f"Sum of list: {sum_list(my_numbers)}")

#Factorial calculator
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print("\n")
print(f"Factorial of 5: {factorial(5)}")
print(f"Factorial of 7: {factorial(7)}")

#Simple calculator
def calculator():
    print("\n--- Simple Calculator ---")
    print("Operations: +, -, *, /")

    num1 = float(input("Enter first number: "))
    op = input("Enter operation (+, -, *, /): ")
    num2 = float(input("Enter second number: "))

    if op == '+':
        result = num1 + num2
    elif op == '-':
        result = num1 - num2
    elif op == '*':
        result = num1 * num2
    elif op == '/':
        if num2 != 0:
            result = num1 / num2
        else:
            result = "Error: Division by zero!"
    else:
        result = "Invalid operation!"

    print(f"Result: {result}")

calculator()
```
